```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
```

```python
def map(x, x_min, x_max, y_min, y_max):
    (y_max-y_min)/(x_max-x_min) * (x-x_min) + y_min

def linspace(start, stop, steps):
    num = start
    step = (stop - start) / steps
    while num < stop:
        yield num
        num += step

def freq(res,cap):
    return 1 / (1.4 * res * cap)

def r1(x):
    return 1e6 * x

def r2(x):
    return 1e6 * (1-x)

def Rexp(x, r):
    return (r1(x) * r / (r1(x)+r))

def logPot(travel):
    #http://benholmes.co.uk/posts/2017/11/logarithmic-potentiometer-laws
    # Define the factor of resistance at the middle pot position
    mid = 0.15
    # b is the base of the exponential equation
    b = (1/mid - 1)**2
    #a is the multiple and also the offset
    a = 1/(b-1)
    # Calculate the resistance
    return a*(b)**travel - a

r_pot = 1e6
    
r_pot_lin = [r_pot * x for x in linspace(0, 1, 280)]
r_pot_log = [r_pot * logPot(t) for t in linspace(0, 1, 280)]

c = 4.7e-6
r0 = 10e3

freqs_lin = [freq(r0+r, c) for r in r_pot_lin]
freqs_log = [freq(r0+r, c) for r in r_pot_log]

fig, ax = plt.subplots(2,2, figsize=(9,6))
fig.tight_layout(pad=3)

#ax[0].plot(range(281), freqs_log)
ax[0][0].plot(range(281), r_pot_lin)
ax[0][1].plot(range(281), freqs_lin)

ax[0][0].grid()
ax[0][0].set_xlabel("Poti Angle")
ax[0][0].set_ylabel("Resistance")
ax[0][0].set_title("Lin. Poti")

ax[0][1].grid()
ax[0][1].set_xlabel("Poti Angle")
ax[0][1].set_ylabel("Frequency")
ax[0][1].set_title("555-Timer Freq.")

ax[1][0].plot(range(281), r_pot_log)
ax[1][1].plot(range(281), freqs_log)

ax[1][0].grid()
ax[1][0].set_xlabel("Poti Angle")
ax[1][0].set_ylabel("Resistance")
ax[1][0].set_title("Log. Poti")

ax[1][1].grid()
ax[1][1].set_xlabel("Poti Angle")
ax[1][1].set_ylabel("Frequency")
ax[1][1].set_title("555-Timer Freq.")

print(freq(r0+1e6,c))

plt.savefig("lin_vs_log.png")
```
